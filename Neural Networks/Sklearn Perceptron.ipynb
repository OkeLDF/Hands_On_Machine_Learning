# Imports and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
iris = load_iris()
X = iris['data']
y = iris['target']
```

```python
iris.target_names
```

```python
# X = X[:, (2,3)] # comprimento e largura da pÃ©tala
X
```

```python
y = (y == 0).astype(np.int16) # setosa: True || False
y 
```

```python
X_data = pd.DataFrame(X, columns=iris.feature_names)
X_data.info()
```

```python
y_test = [5, 3.1, 2, 0.5]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=[10,10])

ind = [[0,1], [2,3], [0,2], [1,3]]
j = 0

for i in range(4):
    row, col = i//2, i%2
    ax[row,col].set_title(f'{X_data.columns[ind[j][0]]} x {X_data.columns[ind[j][1]]}')
    
    ax[row,col].scatter(
        X_data[X_data.columns[ind[j][0]]],
        X_data[X_data.columns[ind[j][1]]],
        c=y, cmap='rainbow'
    )
    
    ax[row,col].scatter(
        [y_test[ind[j][0]]],
        [y_test[ind[j][1]]],
        c='black', marker='x'
    )
    
    j += 1
plt.show()
```

# Perceptron

```python
per_clf = Perceptron()
```

```python
per_clf.fit(X, y)
```

```python
y_pred = per_clf.predict([y_test])
```

```python
print(' Instance :', y_test)
print('Predicted :', y_pred[0])
print('    Class :', (True if y_pred[0]==1 else False), 'setosa')
```
